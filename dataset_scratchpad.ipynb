```python
def _pre_process_text(filepath):
        with open (filepath, 'r', encoding="utf-8") as f:
            text = f.read()
        text = re.sub('\n', ' ', text) # get rid of new lines
        text = text.lower() # make everything lowercase
        text = sent_tokenize(text) # break text into sentences
        # sentence_tok_text = [word_tokenize(word) for word in sentence_tok_text]
        sentence_tok_text =[]
        for sentence in text:
            sentence = f"<BOS>{sentence}<EOS>" # add <BOS> <EOS> tokens
            sentence = re.sub(r'[^\w<>]+', ' ', sentence) # regex to remove punctuation except token markers
            sentence_tok_text.append(sentence)
        return sentence_tok_text # return list of sentences

foo = _pre_process_text(os.path.join('dataset', 'Can_You_Forgive_Her_test_paras copy.txt'))
# foo
print(foo)
print(foo[4])
```

```python
# tokenize the input string and convert it to a tensor
input_sentence = "Vavasor was a man entirely without good qualities"

input_sentence = input_sentence.lower()
input_sentence = re.sub(r'[^\w<>]+', ' ', input_sentence) # regex to remove punctuation except token markers
print(input_sentence)
input_sentence = word_tokenize(input_sentence)
print(input_sentence)
input_sentence.append('<EOS>')
input_sentence.insert(0, '<BOS>')#f"<BOS> {input_sentence} <EOS>" # add <BOS> <EOS> tokens with a space
print(input_sentence)

input_phrase = [vocab.index(w) for w in input_sentence]

print(input_phrase)

input_tensor = torch.tensor(input_phrase, dtype=torch.long).unsqueeze(0).to(device)

print(input_tensor)
```
